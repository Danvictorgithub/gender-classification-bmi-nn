```python
import tensorflow as tf
from keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import csv
```

```python
path = Path('./bmi.csv')
lines = path.read_text().splitlines()
reader = csv.reader(lines)
header_row = next(reader)

for i,row in enumerate(header_row):
    print(i,row)
```

```python
m = 0
n = len(header_row)

x_data_list,y_data_list = [],[]
for row in reader:
    x_data = [float(i) for i in row[1:n]]
    y_data = 0. if row[0] == 'Male' else 1.
    x_data_list.append(x_data)
    y_data_list.append(y_data)
print(x_data_list)
print(y_data_list)
```

```python
x_data_array = np.array(x_data_list)
y_data_array = np.array(y_data_list)

# Create a scatter plot with gender as color and index as size
plt.figure(figsize=(10, 6))
scatter = plt.scatter(x_data_array[:, 0], x_data_array[:, 1], c=y_data_array, s=x_data_array[:, 2]*10, cmap='bwr', alpha=0.5)
plt.colorbar(scatter, label='Gender (0: Male, 1: Female)')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight by Gender with Index as Size')
plt.show()

# Pair Plot (using matplotlib)
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
scatter = plt.scatter(x_data_array[:, 0], x_data_array[:, 1], c=y_data_array, cmap='bwr', alpha=0.5)
plt.colorbar(scatter, label='Gender (0: Male, 1: Female)')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')

plt.subplot(1, 3, 2)
scatter = plt.scatter(x_data_array[:, 0], x_data_array[:, 2], c=y_data_array, cmap='bwr', alpha=0.5)
plt.colorbar(scatter, label='Gender (0: Male, 1: Female)')
plt.xlabel('Height')
plt.ylabel('Index')
plt.title('Height vs Index')

plt.subplot(1, 3, 3)
scatter = plt.scatter(x_data_array[:, 1], x_data_array[:, 2], c=y_data_array, cmap='bwr', alpha=0.5)
plt.colorbar(scatter, label='Gender (0: Male, 1: Female)')
plt.xlabel('Weight')
plt.ylabel('Index')
plt.title('Weight vs Index')

plt.tight_layout()
plt.show()
```
